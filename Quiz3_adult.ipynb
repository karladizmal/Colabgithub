```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Egade/evidences-of-activities-karladizmal/main/adult.csv?token=GHSAT0AAAAAAB5NVJ3FVWL2P2MFEAMQ2IBSY64OK7A')
df.head()
```

```python
x = pd.pivot_table(df,
                   index = 'marital-status',
                   aggfunc = 'count',
                   values = 'sex',
                   )
x
```

```python
x = pd.pivot_table(df,
                   index = ['education', 'marital-status'],
                   aggfunc = 'median',
                   values = ['age', 'hours-per-week'],
                   columns = 'sex'
                   )
x
```

```python
x = pd.pivot_table(df,
                   index = ['education', 'marital-status'],
                   aggfunc = 'mean',
                   values = ['age', 'hours-per-week'],
                   )
x
```
