```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Egade/evidences-of-activities-karladizmal/main/antropometria-dataset-2.csv?token=GHSAT0AAAAAAB5NVJ3FU6DSQ3BJLADA6FVWY64KXYQ')
df.head()
```

```python
x = pd.pivot_table(df,
                   index = 'desc_ent',
                   aggfunc = 'mean',
                   values = 'peso',
                   )
x
```

```python
x = pd.pivot_table(df,
                   index = 'desc_ent',
                   aggfunc = 'mean',
                   values = 'peso',
                   columns = 'sexo'
                   )
x
```
